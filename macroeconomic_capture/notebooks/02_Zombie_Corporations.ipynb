We quantitatively isolate zombie corporations sustained by rent-seeking interventionism, algorithmically differentiating them from economic clusters subject to pure **creative destruction**.*

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
import warnings
warnings.filterwarnings('ignore')

plt.style.use('dark_background')
sns.set_palette("viridis")
```

```python
# Loading structured data from institutional capture
df_corp = pd.read_csv('../data/corporate_zombies_synthetic.csv')
df_corp.head()
```

### 1. Unsupervised Topological Anomaly Detection (Zombification)
We will use a density-based machine learning model (DBSCAN) to dynamically classify which corporations find themselves in a region of permanent insolvency (highly leveraged, low ICR) but survive anomalously (high state subsidy intake).

```python
features = ['Interest_Coverage_Ratio', 'Debt_to_Equity', 'State_Subsidies_Millions']
X = df_corp[features]

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# DBSCAN detects clusters and isolates noise/topological anomalies generated by subsidies
dbscan = DBSCAN(eps=0.8, min_samples=20)
df_corp['ML_Cluster'] = dbscan.fit_predict(X_scaled)

print("Cluster Distribution (-1 indicates subsidy-generated noise/topological anomaly):")
print(df_corp['ML_Cluster'].value_counts())
```

### 2. Creative Destruction vs Institutional Capture
Let's visualize how the productive sector separates from the anomalous companies artificially sustained.

```python
fig, ax = plt.subplots(figsize=(12, 8))

scatter = ax.scatter(
    df_corp['Interest_Coverage_Ratio'], 
    df_corp['State_Subsidies_Millions'], 
    c=df_corp['Is_Zombie'], # True classification label from generator
    cmap='coolwarm', 
    alpha=0.6, 
    s=df_corp['Debt_to_Equity']*20
)

ax.axvline(x=1.0, color='yellow', linestyle='--', alpha=0.5, label='Insolvency Boundary (ICR = 1)')
ax.set_title('Rent-Seeking Topology: State Subsidies vs Solvency (ICR)', fontsize=14)
ax.set_xlabel('Interest Coverage Ratio (EBIT / Interest)')
ax.set_ylabel('Received State Subsidies ($M)')

cbar = plt.colorbar(scatter)
cbar.set_label('Zombie Condition Probability (Red = High)')

plt.legend()
plt.show()
```

### Paretian Conclusion
Companies located in the upper left quadrant (ICR < 1, high subsidies) are the **Zombie Corporations**. They consume scarce capital avoiding their technical bankruptcy, inhibiting *Schumpeterian Creative Destruction* and reducing the aggregate productivity of the economic ecosystem.
