*Nota Institucional: Este modelo demuestra el cruce topológico de la distribución del déficit público con las tasas de rentabilidad sectorial. Aislamos cuantitativamente las corporaciones zombis sostenidas por el intervencionismo rentista, diferenciándolas algorítmicamente de los clústeres económicos sujetos a **destrucción creativa** pura.*

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
import warnings
warnings.filterwarnings('ignore')

plt.style.use('dark_background')
sns.set_palette("viridis")
```

```python
# Carga de datos estructurados de captura institucional
df_corp = pd.read_csv('../data/corporate_zombies_synthetic.csv')
df_corp.head()
```

### 1. Detección Topológica No Supervisada de Anomalías (Zombificación)
Utilizaremos un modelo de machine learning basado en densidad (DBSCAN) para clasificar dinámicamente qué corporaciones se encuentran en una región de insolvencia permanente (altamente apalancadas, bajo ICR) pero que sobreviven de manera anómala (alta ingesta de subsidios estatales).

```python
features = ['Interest_Coverage_Ratio', 'Debt_to_Equity', 'State_Subsidies_Millions']
X = df_corp[features]

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# DBSCAN detects clusters and isolates noise/anomalies
dbscan = DBSCAN(eps=0.8, min_samples=20)
df_corp['ML_Cluster'] = dbscan.fit_predict(X_scaled)

print("Distribución de Clústeres (-1 indica ruido/anomalía topológica generada por subsidios):")
print(df_corp['ML_Cluster'].value_counts())
```

### 2. Destrucción Creativa vs Captura Institucional
Visualicemos cómo el sector productivo se separa de las empresas anómalas sostenidas artificialmente.

```python
fig, ax = plt.subplots(figsize=(12, 8))

scatter = ax.scatter(
    df_corp['Interest_Coverage_Ratio'], 
    df_corp['State_Subsidies_Millions'], 
    c=df_corp['Is_Zombie'], # True classification label from generator
    cmap='coolwarm', 
    alpha=0.6, 
    s=df_corp['Debt_to_Equity']*20
)

ax.axvline(x=1.0, color='yellow', linestyle='--', alpha=0.5, label='Frontera de Insolvencia (ICR = 1)')
ax.set_title('Topología de Rentismo: Subsidios Estatales vs Solvencia (ICR)', fontsize=14)
ax.set_xlabel('Interest Coverage Ratio (EBIT / Interest)')
ax.set_ylabel('Subsidios Estatales Recibidos ($M)')

cbar = plt.colorbar(scatter)
cbar.set_label('Probabilidad de Condición Zombi (Rojo = Alta)')

plt.legend()
plt.show()
```

### Conclusión Paretiana
Las empresas ubicadas en el cuadrante superior izquierdo (ICR < 1, altos subsidios) son las **Corporaciones Zombis**. Consumen capital escaso evitando su quiebra técnica, inhibiendo la *Destrucción Creativa Schumpeteriana* y reduciendo la productividad agregada del ecosistema económico.
