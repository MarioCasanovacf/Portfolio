*Nota Institucional: Este notebook operacionaliza la fenomenología hegeliana reduciendo su corpus textual a una topología de red determinista. Mediante la extracción de tripletas semánticas (Sujeto-Predicado-Objeto), se modela el flujo de la 'Aufhebung' (sublimación) como aristas direccionales convergentes. El análisis culmina aplicando algoritmos de Centralidad de Autovectores (Eigenvector Centrality) cuantificando el peso empírico de las categorías lógicas dentro de la arquitectura del sistema.*

```python
import numpy as np
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')

plt.style.use('dark_background')
```

### 1. Ingestión del Corpus y Extracción de Tripletas
Se simula la extracción de NLP (Procesamiento de Lenguaje Natural) parseando las afirmaciones lógicas en formato relacional estructurado legible por máquina.

```python
# Corpus Textual (Ingestión simulada)
"""
El Ser en sí mismo es puro e indeterminado, pero en su inmediatez inevitablemente 
transiciona hacia la Nada. Esta inmediatez del Ser y la Nada encuentra su verdad en 
el Devenir. De igual manera, el Amo arriesga su vida por el reconocimiento, estableciendo el 
paradigma del Dominio. El Esclavo rehuye la muerte y abraza el Trabajo, transformando 
la Naturaleza objetiva. El Trabajo del Esclavo trasciende el Dominio estancado del Amo. 
A través de la Aufhebung, el Trabajo y el Dominio convergen en el Autoconocimiento absoluto. 
La Razón supera al Autoconocimiento subjetivo para alcanzar el Espíritu.
"""

# Extracción Estructurada de Tripletas Orgánicas (Sujeto, Predicado, Objeto)
triplets = [
    ("Ser", "transiciona_hacia", "Nada"),
    ("Ser", "confluye_en", "Devenir"),
    ("Nada", "confluye_en", "Devenir"),
    ("Amo", "establece", "Dominio"),
    ("Esclavo", "abraza", "Trabajo"),
    ("Esclavo", "transforma", "Naturaleza"),
    ("Trabajo", "trasciende", "Dominio"),
    ("Trabajo", "aufhebung_hacia", "Autoconocimiento"),
    ("Dominio", "aufhebung_hacia", "Autoconocimiento"),
    ("Razón", "supera", "Autoconocimiento"),
    ("Razón", "alcanza", "Espíritu")
]

df_triplets = pd.DataFrame(triplets, columns=["Sujeto", "Predicado", "Objeto"])
df_triplets
```

### 2. Construcción del Grafo Direccional (Topología de Red)
Transformamos las tuplas relacionales en un Grafo Dirigido ($G=(V,E)$) utilizando `NetworkX` para auditar la direccionalidad convergente de los nodos contradictorios.

```python
G = nx.DiGraph()

# Añadir aristas vectoriales basadas en las tripletas
for _, row in df_triplets.iterrows():
    G.add_edge(row['Sujeto'], row['Objeto'], relation=row['Predicado'])

print(f"[*] Topología de Red: {G.number_of_nodes()} Nodos Dialécticos, {G.number_of_edges()} Aristas Lógicas.")
```

### 3. Álgebra Numérica de Grafos: Eigenvector Centrality
La centralidad de autovectores ($Ax = \lambda x$) mide no sólo cuántas conexiones tiene un nodo, sino la *influencia paramétrica* de los nodos que apuntan a él. Aplicado a la fenomenología, cuantifica matemáticamente el peso sistémico de cada estadio lógico al converger histórico-conceptualmente (Aufhebung).

```python
# Cálculo Numérico de Centralidad de Autovectores (Iterando el método de la potencia)
eigen_centrality = nx.eigenvector_centrality(G, max_iter=1000, tol=1e-06)

# Normalización a escala de Carga Lógica (0 a 100)
df_centrality = pd.DataFrame(list(eigen_centrality.items()), columns=['Categoría_Lógica', 'Peso_Vectorial'])
df_centrality['Peso_Vectorial'] = (df_centrality['Peso_Vectorial'] / df_centrality['Peso_Vectorial'].max()) * 100

df_centrality.sort_values(by='Peso_Vectorial', ascending=False).reset_index(drop=True)
```

### 4. Visualización Semántica de la Red
Renderizamos la estructura donde el radio de los nodos ($r$) es estocásticamente proporcional a su vector de eigen-centralidad, probando empíricamente que *Devenir* y *Autoconocimiento* absorben la energía de la contradicción axiomática.

```python
plt.figure(figsize=(14, 10))

# Usar escalado de autovectores para el tamaño nodal
node_sizes = [eigen_centrality[node] * 6000 for node in G.nodes()]

# Algoritmo de posicionamiento Kamada-Kawai (óptimo para despliegues de fuerza dirigida)
pos = nx.kamada_kawai_layout(G)

nx.draw_networkx_nodes(G, pos, node_size=node_sizes, node_color="magenta", alpha=0.7, edgecolors="white")
nx.draw_networkx_edges(G, pos, edge_color="cyan", arrowsize=20, arrowstyle="->", width=1.5, alpha=0.6)
nx.draw_networkx_labels(G, pos, font_size=11, font_color="white", font_weight="bold")

# Etiquetas de predicados relacionales
edge_labels = nx.get_edge_attributes(G, 'relation')
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_color="yellow", font_size=9, label_pos=0.5)

plt.title("Grafo de Conocimiento Dialéctico (Aufhebung y Eigen-Centralidad)", fontsize=16, pad=20, color="white")
plt.axis("off")
plt.show()
```
