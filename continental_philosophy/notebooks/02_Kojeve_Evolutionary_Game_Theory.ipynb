*Nota Institucional: Este ecosistema estocástico somete la dialéctica del reconocimiento histórico (Alexandre Kojève) a un modelo de agentes autónomos. Operando bajo matrices de pagos asimétricas continuas —donde la utilidad marginal de cada agente está dictada por su "deseo de reconocimiento" frente al "miedo a la muerte"— el sistema abstrae axiomas morales en topología termodinámica. El motor audita la bifurcación predecible de clases y registra empíricamente el agotamiento del sistema hacia el Estado Homogéneo Universal.*

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

plt.style.use('dark_background')
sns.set_palette("flare")
np.random.seed(42)
```

### 1. Ingestión de Agentes y Entorno Estocástico
Inyectamos el set pre-generado de 1000 agentes poblacionales portando variables continuas pre-asignadas.

```python
df_agents = pd.read_csv("../data/kojeve_initial_agents.csv")
n_agents = len(df_agents)

# Inicializamos cada agente en un estado de Consciencia Autocerrada (Neutral)
df_agents['Clase_Topologica'] = 'Neutral'
df_agents['Energia_Historica'] = 100.0  # Capacidad de iteración

print(f"[*] Simulación cargada: {n_agents} Agentes Autónomos.")
df_agents.head()
```

### 2. Matriz de Pagos Asimétrica (Mecánica Subyacente)
La lucha por el reconocimiento rige que al colisionar dos agentes:
- Si ambos ceden al miedo: Resultan sobrevivientes paralizados.
- Si uno cede y el otro avanza: Se bifurca la topología en Clase Productiva (Esclavo/Trabajador) y Clase Extractiva (Amo/Consumidor).
- Si ninguno cede: Extinción o pérdida dramática de Energía Histórica.

```python
def simular_lucha_reconocimiento(agente_A, agente_B):
    """
    Evalúa el choque de consciencias basado en umbrales estocásticos.
    Retorna los cambios de clase resultantes para [A, B].
    """
    # ¿Avanzan a la lucha o ceden al miedo?
    # Es función del deseo de reconocimiento superando al miedo a la muerte
    A_avanza = agente_A['Deseo_de_Reconocimiento'] > agente_A['Miedo_a_la_Muerte']
    B_avanza = agente_B['Deseo_de_Reconocimiento'] > agente_B['Miedo_a_la_Muerte']
    
    if A_avanza and B_avanza:
        # Choque fatal, desgaste brutal para ambos sin sublimación
        return 'Extinto', 'Extinto'
    elif A_avanza and not B_avanza:
        # A subyuga a B
        return 'Extractiva', 'Productiva'
    elif not A_avanza and B_avanza:
        # B subyuga a A
        return 'Productiva', 'Extractiva'
    else:
        # Ambos se retraen, se mantienen Neutrales
        return 'Neutral', 'Neutral'

# Simulación Monte Carlo del Torneo Inicial (10 Iteraciones en red)
agent_records = df_agents.to_dict('records')

historicos_clase = []

# Motor Evolutivo
for epoca in range(15):
    # Mezcla estocástica de emparejamientos en la arena temporal
    np.random.shuffle(agent_records)
    
    # Enfrentamientos Pareados
    for i in range(0, n_agents, 2):
        if agent_records[i]['Clase_Topologica'] == 'Neutral' and agent_records[i+1]['Clase_Topologica'] == 'Neutral':
            res_A, res_B = simular_lucha_reconocimiento(agent_records[i], agent_records[i+1])
            agent_records[i]['Clase_Topologica'] = res_A
            agent_records[i+1]['Clase_Topologica'] = res_B

        # La clase Productiva sublima (Aufhebung) a la Extractiva a lo largo de las épocas.
        # Agotamiento de la asimetría.
        elif agent_records[i]['Clase_Topologica'] == 'Extractiva' and agent_records[i+1]['Clase_Topologica'] == 'Productiva':
            # La creatividad productiva transforma la asimetría en un ciudadano homogéneo (Igualdad final)
            if np.random.rand() > 0.7:  # Probabilidad de sublimación temporal
                agent_records[i]['Clase_Topologica'] = 'Ciudadano Homogéneo'
                agent_records[i+1]['Clase_Topologica'] = 'Ciudadano Homogéneo'
        elif agent_records[i]['Clase_Topologica'] == 'Productiva' and agent_records[i+1]['Clase_Topologica'] == 'Extractiva':
             if np.random.rand() > 0.7:
                agent_records[i]['Clase_Topologica'] = 'Ciudadano Homogéneo'
                agent_records[i+1]['Clase_Topologica'] = 'Ciudadano Homogéneo'
                
    # Captura analítica Columnar del Corte transversal de tiempo
    df_temp = pd.DataFrame(agent_records)
    dist_clase = df_temp['Clase_Topologica'].value_counts().to_dict()
    dist_clase['Epoca'] = epoca
    historicos_clase.append(dist_clase)

df_historia = pd.DataFrame(historicos_clase).fillna(0)
```

### 3. Termodinámica Demográfica: El Colapso al Estado Universal
En lugar de postular una teleología moral, el modelo de agentes demuestra algebraicamente que la acumulación asimétrica es topológicamente inestable en un entorno iterativo de juego infinito cerrado. La entropía del conflicto colapsa irremediablemente hacia su sumidero estacionario de máxima inercia: el **Estado Homogéneo Universal**.

```python
plt.figure(figsize=(12, 7))

if 'Neutral' in df_historia.columns:
    plt.plot(df_historia['Epoca'], df_historia['Neutral'], color='gray', linestyle=':', label='Consciencias Neutrales (Inactivas)')
if 'Extractiva' in df_historia.columns:
    plt.plot(df_historia['Epoca'], df_historia['Extractiva'], color='red', linewidth=2.5, label='Topología Extractiva (Amo)')
if 'Productiva' in df_historia.columns:
    plt.plot(df_historia['Epoca'], df_historia['Productiva'], color='orange', linewidth=2.5, label='Topología Productiva (Trabajador)')
if 'Ciudadano Homogéneo' in df_historia.columns:
    plt.plot(df_historia['Epoca'], df_historia['Ciudadano Homogéneo'], color='cyan', linewidth=3, linestyle='--', label='Saturación: Estado Homogéneo Universal')
if 'Extinto' in df_historia.columns:
    plt.plot(df_historia['Epoca'], df_historia['Extinto'], color='black', label='Extinción Paramétrica', alpha=0.5)

plt.fill_between(df_historia['Epoca'], df_historia.get('Extractiva', 0), color='red', alpha=0.1)
plt.fill_between(df_historia['Epoca'], df_historia.get('Ciudadano Homogéneo', 0), color='cyan', alpha=0.1)

plt.title("Transición de Fase de la Dialéctica Kojèviana (Teoría de Juegos Evolutiva)", fontsize=16, pad=20)
plt.xlabel("Iteración Estocástica (Tiempo Histórico)", fontsize=12)
plt.ylabel("Densidad Nodal (Volumen de Agentes)", fontsize=12)
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.grid(alpha=0.15)

plt.show()
```
