*Note: This stochastic ecosystem subjects the dialectic of historical recognition (Alexandre Kojève) to an autonomous agent model. Operating under continuous asymmetric payoff matrices —where each agent's marginal utility is dictated by its "desire for recognition" versus "fear of death"— the system abstracts moral axioms into thermodynamic topology. The engine audits the predictable bifurcation of classes and empirically records the system's exhaustion toward the Universal Homogeneous State.*

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

plt.style.use('dark_background')
sns.set_palette("flare")
np.random.seed(42)
```

### 1. Agent Ingestion and Stochastic Environment
We inject the pre-generated set of 1,000 population agents carrying pre-assigned continuous variables.

```python
df_agents = pd.read_csv("../data/kojeve_initial_agents.csv")
n_agents = len(df_agents)

# We initialize each agent in a state of Self-closed Consciousness (Neutral)
df_agents['Topological_Class'] = 'Neutral'
df_agents['Historical_Energy'] = 100.0  # Iteration capacity

print(f"[*] Simulation loaded: {n_agents} Autonomous Agents.")
df_agents.head()
```

### 2. Asymmetric Payoff Matrix (Underlying Mechanics)
The struggle for recognition dictates that when two agents collide:
- If both yield to fear: They result in paralyzed survivors.
- If one yields and the other advances: The topology bifurcates into a Productive Class (Slave/Worker) and an Extractive Class (Master/Consumer).
- If neither yields: Extinction or dramatic loss of Historical Energy.

```python
def simulate_struggle_for_recognition(agent_A, agent_B):
    """
    Evaluates the clash of consciousness based on stochastic thresholds.
    Returns the resulting class changes for [A, B].
    """
    # Do they advance to the struggle or yield to fear?
    # It is a function of the desire for recognition overcoming the fear of death
    A_advances = agent_A['Desire_for_Recognition'] > agent_A['Fear_of_Death']
    B_advances = agent_B['Desire_for_Recognition'] > agent_B['Fear_of_Death']
    
    if A_advances and B_advances:
        # Fatal clash, brutal exhaustion for both without sublation
        return 'Extinct', 'Extinct'
    elif A_advances and not B_advances:
        # A subjugates B
        return 'Extractive', 'Productive'
    elif not A_advances and B_advances:
        # B subjugates A
        return 'Productive', 'Extractive'
    else:
        # Both retract, remain Neutral
        return 'Neutral', 'Neutral'

# Monte Carlo Simulation of the Initial Tournament (15 Network Iterations)
agent_records = df_agents.to_dict('records')

class_history = []

# Evolutionary Engine
for epoch in range(15):
    # Stochastic shuffling of pairings in the temporal arena
    np.random.shuffle(agent_records)
    
    # Paired Battles
    for i in range(0, n_agents, 2):
        if agent_records[i]['Topological_Class'] == 'Neutral' and agent_records[i+1]['Topological_Class'] == 'Neutral':
            res_A, res_B = simulate_struggle_for_recognition(agent_records[i], agent_records[i+1])
            agent_records[i]['Topological_Class'] = res_A
            agent_records[i+1]['Topological_Class'] = res_B

        # The Productive class sublates (Aufhebung) the Extractive class over the epochs.
        # Exhaustion of asymmetry.
        elif agent_records[i]['Topological_Class'] == 'Extractive' and agent_records[i+1]['Topological_Class'] == 'Productive':
            # Productive creativity transforms asymmetry into a homogeneous citizen (Final equality)
            if np.random.rand() > 0.7:  # Probability of temporal sublation
                agent_records[i]['Topological_Class'] = 'Universal Citizen'
                agent_records[i+1]['Topological_Class'] = 'Universal Citizen'
        elif agent_records[i]['Topological_Class'] == 'Productive' and agent_records[i+1]['Topological_Class'] == 'Extractive':
             if np.random.rand() > 0.7:
                agent_records[i]['Topological_Class'] = 'Universal Citizen'
                agent_records[i+1]['Topological_Class'] = 'Universal Citizen'
                
    # Columnar analytical capture of the cross-sectional time slice
    df_temp = pd.DataFrame(agent_records)
    class_dist = df_temp['Topological_Class'].value_counts().to_dict()
    class_dist['Epoch'] = epoch
    class_history.append(class_dist)

df_history = pd.DataFrame(class_history).fillna(0)
```

### 3. Demographic Thermodynamics: The Collapse to the Universal State
Instead of positing a moral teleology, the agent model algebraically demonstrates that asymmetric accumulation is topologically unstable in a closed infinite game iterative environment. Conflict entropy irremediably collapses toward its stationary sink of maximum inertia: the **Universal Homogeneous State**.

```python
plt.figure(figsize=(12, 7))

if 'Neutral' in df_history.columns:
    plt.plot(df_history['Epoch'], df_history['Neutral'], color='gray', linestyle=':', label='Neutral Consciousnesses (Inactive)')
if 'Extractive' in df_history.columns:
    plt.plot(df_history['Epoch'], df_history['Extractive'], color='red', linewidth=2.5, label='Extractive Topology (Master)')
if 'Productive' in df_history.columns:
    plt.plot(df_history['Epoch'], df_history['Productive'], color='orange', linewidth=2.5, label='Productive Topology (Worker)')
if 'Universal Citizen' in df_history.columns:
    plt.plot(df_history['Epoch'], df_history['Universal Citizen'], color='cyan', linewidth=3, linestyle='--', label='Saturation: Universal Homogeneous State')
if 'Extinct' in df_history.columns:
    plt.plot(df_history['Epoch'], df_history['Extinct'], color='black', label='Parametric Extinction', alpha=0.5)

plt.fill_between(df_history['Epoch'], df_history.get('Extractive', 0), color='red', alpha=0.1)
plt.fill_between(df_history['Epoch'], df_history.get('Universal Citizen', 0), color='cyan', alpha=0.1)

plt.title("Phase Transition of Kojèvian Dialectics (Evolutionary Game Theory)", fontsize=16, pad=20)
plt.xlabel("Stochastic Iteration (Historical Time)", fontsize=12)
plt.ylabel("Node Density (Volume of Agents)", fontsize=12)
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.grid(alpha=0.15)

plt.show()
```
